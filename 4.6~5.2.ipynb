##4.6정리

*   기계학습에서 사용하는 데이터셋은 훈련 데이터와 시험 데이터로 나눠 사용한다
*   훈련 데이터로 학습한 모델의 범용 능력을 시험 데이터로 평가한다
*   신경망 학습은 손실 함수를 지표로, 손실 함수의 값이 작아지는 방향으로 가중치 매개변수를 갱신한다
*   가중치 매개변수를 갱신할 때는 가중치 매개변수의 기울기를 이용하고, 기울어진 방향으로 가중치의 값을 갱신하는 작업을 반복한다
*   아주 작은 값을 주었을 때의 차분으로 미분하는 것을 수치 미분이라고 한다
*   수치 미분을 이용해 가중치 매개변수의 기울기를 구할 수 있다
*   수치 미분을 이용한 게산에는 시간이 걸리지만, 그 구현은 간단하다. 한편, 오차역전파법은 기울기를 고속으로 구할 수 있다



##5.1계산그래프

*  계산 과정을 그래프로 나타낸 것
*  노드와 에지와 표현됨(에지는 노드사이의 직선임)


**계산 그래프로 풀다**
문제
1.   문제 : 현빈 군은 슈퍼에서 1개에 100원인 사과를 2개 샀습니다. 소비세가 10%일 때 지불 금액을 구하세요

2.  문제 : 현빈 군은 슈퍼에서 사과를 2개, 굴을 3개 샀습니다. 사과는 1개에 100원, 귤은 1개 150원입니다. 소비세가 10%일 때 지불 금액을 구하세요

*   계산 과정을 노드와 화살표로 표시함
*  각 노드의 계산 결과가 왼쪽에소 오른쪽으로 전해지게 됨
*   계산을 왼쪽에서 오른쪽으로 진행하는 단계를 순전파라 함
*   순전파의 반대가 역전파이고 미분계산에서 중요한 역할을 함




**국소적 계산**

*  계산 그래프의 특징:"국소적 계산"을 전파함으로서 최종 결과를 얻는다
*   즉 전체에서 어떤 일이 생기든 자신과 관계된 정보만으로 결과를 출력할 수 있음



**왜 계산 그래프로 푸는가**

*  국소적 계산
*  중간 계산 결과를 모두 보관할 수 있음
*  역전파를 통해 미분을 효율적으로 계산 할 수 있음
*  역전파는 굵은 화살표를 사용함   



##5.2 연쇄법칙
연쇄법칙: 국소적 미분을 전달하는 원리

**계산 그래프의 역전파**

*   역전파의 계산 절차 : 화살표 E에 노드의 국소적 미분을 곱한 후 다음 노드로 전달하는 것 
*  국소적미분 : 순전파 때의 y=f(x) 계산의 미분을 구한다
*  이러한 방식으로 목표로 하는 미분 값을 효율적으로 구할 수 있음





**연쇄법칙이란?**

*   연쇄법칙은 합성함수의 미분에 대한 성질이며 다음과 같이 정리됨 - 합성 함수의 미분은 합성 함수를 구성하는 각 함수의 미분의 곱으로 나타낼 수 있음
*  예) x에 대한 Z의 미분은 t에 대한 z의 미분과 x에 대한 t의 미분의 곱으로 나타낼 수 있음



**연쇄법칙과 계산 그래프**

*   계산 그래프의 역전파는 오른쪽에서 왼쪽으로 신호를 전파함
*   역전파의 계산 절차에서는 노드롸 들어온 입력 신호에 그 노드의 국소적 미분(편미분)을 곱한 후 다음 노드로 전달함
